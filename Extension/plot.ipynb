```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot(title, values):
    min_loss = min(values)
    min_loss_idx = values.index(min_loss)

    # Create the plot
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=range(len(values)), y=values, marker='o', label='Loss')

    # Add a dashed line for the minimum loss
    plt.axhline(y=min_loss, color='red', linestyle='--', label=f'Min Loss: {min_loss:.4f}')
    plt.scatter(min_loss_idx, min_loss, color='red', zorder=5)  # Highlight the minimum point

    # Annotate the minimum point
    plt.text(min_loss_idx, min_loss, f'  Min ({min_loss:.4f})', color='red', verticalalignment='bottom')

    # Configure plot labels and title
    plt.title(title, fontsize=14)
    plt.xlabel("Epoch", fontsize=12)
    plt.ylabel("Loss", fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()
```
